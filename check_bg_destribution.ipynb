```python
from glob import glob
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_first_frame(video_path):
    vidcap = cv2.VideoCapture(video_path)
    success, image = vidcap.read()
    return image
```

```python
def read_crop_file():
    f = open('crop_sizes.txt', 'r')
    files = []
    values = []
    for l in f.readlines():
        ss = l[:-1].split(' ')
        files.append(ss[0])
        values.append(float(ss[1]))
    return files, np.array(values)
```

```python
crops_files, crop_values = read_crop_file()
```

```python
okvideos = np.array(crops_files)[(crop_values > 0.95) * (crop_values < 0.98)]
```

```python
plt.figure(figsize=(12, 8))
plt.hist(crop_values)
plt.show()
```
